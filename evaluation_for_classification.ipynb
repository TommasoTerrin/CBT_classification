```python
from agentic_graph import distorsions_list

import json
import pandas as pd
from sklearn.metrics import f1_score
```

```python

# Carica il dataset JSON
with open("Dataset HF/distortions_test.json", "r", encoding="utf-8") as f:
    data = json.load(f)

# Converti il JSON in un DataFrame pandas
df = pd.DataFrame(data)
df = df[["ori_text", "distortions"]]  # Tieni solo le colonne necessarie

# Solo prime 10 righe per testare tutto
#df= df.head(10)

print(df)
```

```python
# Applica ricerca di distorsioni per tutte le righe
df["predictions"] = df["ori_text"].apply(distorsions_list)

# Vedi risultati
print(df)
```

```python
# Creiamo un set di tutte le distorsioni presenti nel dataset
all_distortions = ["all-or-nothing thinking", "overgeneralization", "mental filter", "mind reading", "fortune-telling", "magnification","emotional reasoning", "should statements", "labeling", "personalization"]

#test
print(all_distortions)
```

```python
# Creiamo le liste binarie per sklearn
y_true = []
y_pred = []

for _, row in df.iterrows():
    true_labels = set(row["distortions"])
    predicted_labels = set(row["predictions"])
    
    y_true.append([1 if d in true_labels else 0 for d in all_distortions])
    y_pred.append([1 if d in predicted_labels else 0 for d in all_distortions])


# Per test
print(y_true)
print(y_pred)
```

```python
# Calcoliamo F1 Score Macro
f1_macro = f1_score(y_true, y_pred, average='macro')
f1_weighted = f1_score(y_true, y_pred, average='weighted')

# Vediamo quanto
print(f1_macro)
print(f1_weighted)
```

```python
###########################################################################################
# TESTS WITHOUT LANGGRAPH
from pydantic import BaseModel, Field
from typing import List, Optional, Literal
from langchain_openai import ChatOpenAI

class Distorsions(BaseModel):
    class Distorsion(BaseModel):
        distorsion: Literal["all-or-nothing thinking", "overgeneralization", "mental filter", "mind reading", "fortune-telling", "magnification","emotional reasoning", "should statements", "labeling", "personalization"]

    distortions: Optional[List[Distorsion]] = Field(description="Distorsions list, element must be unique")
```

```python
# LLM OBJECT FOR TESTING
def create_dist_llm():
    return ChatOpenAI(
        model="gpt-4o",
        temperature=0.2
    ).with_structured_output(Distorsions)

import prompt as pr 

def prompt_test_1(input_text:str):
    prompt= f""" 
    You are a CBT therapist and now need to determine the cognitive distortions of a patient from An input text.
    Input text from patient: 
    '''{input_text} '''

    You can choose from the following 10 cognitive distortions:
    '''["all-or-nothing thinking", "overgeneralization", "mental filter", "mind reading", "fortune-telling",
    "magnification","emotional reasoning", "should statements", "labeling", "personalization"]'''

    Output format: output must be just the list of distorsions you have detected in the text. 
    In the list every element **must be unique**.
    """
    llm= create_dist_llm()
    dist_list=[]
    response= llm.invoke(prompt)
    if response.distortions:
        for dist in response.distortions:
            dist_list.append(dist.distorsion)
    return dist_list

def prompt_test_2(input_text:str):
    prompt= f"""
    You are a CBT therapist and now need to determine the cognitive distortions of a patient from An input text.
    Input text from patient: 
    '''{input_text} '''
    
    To detects the cognitive distortions of the patient, you have to follow the following istructions:
    '''
    {pr.DISTORSIONS}
    '''
    Istructions:
    1. Detects wich are the toughts inside the input text
    2. Detects the cognitive distiotions of the patient, using the istructions.

    Output format: output must be just the list of distorsions you have detected in the text. 
    In the list every element **must be unique**.
    """
    llm= create_dist_llm()
    dist_list=[]
    response= llm.invoke(prompt)
    if response.distortions:
        for dist in response.distortions:
            dist_list.append(dist.distorsion)
    return dist_list


#Try these two previous methods
res1= f""" Original Text: {df.iloc[0]["ori_text"]} \n
            Distorsions: {df.iloc[0]["distortions"]}\n\n
            -----------------------------------------------
            Predicted Distorsions 1: {prompt_test_1(df.iloc[0]["ori_text"])}\n
            Predicted Distorsions 2: {prompt_test_2(df.iloc[0]["ori_text"])}"""
res2= f"""
    Original Text: {df.iloc[1]["ori_text"]} \n
    Distorsions: {df.iloc[1]["distortions"]}\n\n
    -----------------------------------------------
    Predicted Distorsions: {prompt_test_1(df.iloc[1]["ori_text"])}\n 
    Predicted Distorsions: {prompt_test_2(df.iloc[1]["ori_text"])}"""


print(res1)
print("\n ££££££££££££££££££££££££££££££££££££££££££££ \n")
print(res2)
```
